## **Problema de Negócio**

Sistema de Recomendação de Livros da Amazon

```python
# Importando Pacotes

import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
```

```python
# Impotação dos Dados Referentes aos Usuaários

users = pd.read_csv("/BX-Users.csv", sep=';', encoding='latin-1')
display(users)
```

```python
# Impotação dos Dados Referentes aos livros

books = pd.read_csv("/BX-Books.csv", sep=';', encoding='latin-1',error_bad_lines=False)
display(books)
```

```python
# Impotação dos Dados Referentes aos Ratings

ratings = pd.read_csv("/BX-Book-Ratings.csv", sep=';', encoding='latin-1')

display(ratings)
```

### **Pré-Processamento de Dados**


```python
books
```

```python
# Renomear Colunas

books = books[['ISBN','Book-Title', 'Book-Author','Year-Of-Publication','Publisher']]
books.rename(columns= {'Book-Title':'title','Book-Author':'author','Year-Of-Publication':'year','Publisher':'publisher'},inplace=True)
users.rename(columns = {'User-ID':'user_id','Location':'location','Age': 'age'},inplace=True)
ratings.rename(columns ={'User-ID':'user_id', 'Book-Rating':'rating'},inplace=True )
```

```python
# Quantidade de Ratings por Usuários

ratings['user_id'].value_counts()
```

```python
# Livros que tenham mais de 200 avaliações

x =ratings['user_id'].value_counts()>200
```

```python
# Quantidade Usuários

y = x[x].index
print(y.shape)
```

## **Decisão de Negócio**

```python
# Trazendo ratings somente dos usuários que avaliaram mais de 200 livros

ratings = ratings[ratings['user_id'].isin(y)]
```

```python
# Unindo Tbelas

rating_with_books = ratings.merge(books, on='ISBN')
rating_with_books.head() 
```

```python
# Quantidade de vezes que um livro apareceu em umrating

number_rating = rating_with_books.groupby("title")['rating'].count().reset_index()
display(number_rating)
```

```python
# Renomeando coluna

number_rating.rename(columns={"rating": "number_of_ratings"},inplace=True)
```

```python
# Criar uma tabela relacionada com o rating
final_rating = rating_with_books.merge(number_rating, on= "title")
final_rating
```

### *** Decisão de Negócio***

```python
# Filtrar livros com mais de 50 avaliações

final_rating = final_rating[final_rating['number_of_ratings']>=50]
final_rating.shape
```

```python
# Descartar valores duplicados

final_rating.drop_duplicates(['user_id','title'],inplace=True)
final_rating.shape
```

```python
# Transposição de Linhas em Colunas
book_pivot = final_rating.pivot_table(columns="user_id", index='title', values='rating')
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
book_sparse
```

```python
  from sklearn.neighbors import NearestNeighbors
  model = NearestNeighbors(algorithm = 'brute')
  model.fit(book_sparse)
```

```python
# Livro 1984

distances, suggestions = model.kneighbors(book_pivot.iloc[0,:].values.reshape(1,-1))
```

```python
from os import listxattr
lista_sugestoes = list()

for i in range(len(suggestions)):
  #print(book_pivot.index[suggestions[i]])
  lista_sugestoes.append(book_pivot.index[suggestions[i]])
  print(lista_sugestoes)
```
