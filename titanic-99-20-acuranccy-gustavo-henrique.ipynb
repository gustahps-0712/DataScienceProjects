# # Análise Exploratória na Tabela Train

```python
# Importação da tabela train

train =  pd.read_csv('/kaggle/input/titanic/train.csv')
display(train)
```

```python
# Viualização da Informação da Tabela train

train.info
```

```python
# Viualização da Descrição da Tabela train

train.describe
```

```python
# Verificando se há colunas com itens vazios da Tabela train

train.isnull().sum()
```

```python
# Excluindo a coluna Age e Cabin do modelo da Tabela train

train = train.drop(["Age", "Cabin"],axis=1)
#display(train)
train.isnull().sum()
```

```python
# Exluindo Linhas Vazias da Tabela train

train = train.dropna(how="any",axis=0)
train.isnull().sum()
```

```python
train["Embarked"]= train["Embarked"].replace({"C":1,"Q":2,"S":3})
display(train)
```

```python
train["Sex"]= train["Sex"].replace({"female":1,"male":2})
display(train)
```

# Análise Exploratória na Tabela Test

```python
# Importação da tabela Test

test =  pd.read_csv('/kaggle/input/titanic/test.csv')
display(test)
```

```python
# Viualização da Informação da Tabela Test

test.shape
```

```python
# Viualização da descrição da Tabela Test

test.describe
```

```python
# Verificando Valores Vazios a Tabela Test

test.isnull().sum()
```

```python
#Excluindo Culunas Age e Cabin 

test= test.drop(["Age","Cabin","Ticket"],axis=1)
test.isnull().sum()
```

```python
test=test.dropna(how="any",axis=0)
test.isnull().sum()
```

```python
#Substituindo valores de Letras para valores numéricos

test["Embarked"].replace({"C":1,"Q":2,"S":3},inplace=True)
test["Sex"]=test["Sex"].apply(lambda x: 1 if x == "female" else 2)

display(test)
```

```python
# Correlação entre as variáveis numéricas
test.corr()
```

```python
# Relacionando a tabela teste com a tabela gender

testot = pd.merge(test,gender,on="PassengerId", how= "inner")
display(testot)
```

```python
features = ["Pclass","Sex","SibSp","Parch"]

y_train=train["Survived"]
y_test = testot["Survived"]

x_train= train[features]
x_test = testot[features]

```

```python
# Performance sobre o metodo Floresta Aleatória

modelo = RandomForestClassifier(n_estimators=100, max_depth=100, random_state=1)
modelo = modelo.fit(x_train,y_train)
previsoes = modelo.predict(x_test)


# Verificar a performance do modelo
acuracia = accuracy_score(y_test, previsoes)
print(f"Acurácia da floresta aleatória foi de: {acuracia*100:.2f}%")

```

```python
# Performance sobre o metodo Regressão Logística

modelo = LogisticRegression()
modelo = modelo.fit(x_train,y_train)
previsoes = modelo.predict(x_test)


# Verificar a performance do modelo
acuracia = accuracy_score(y_test, previsoes)
print(f"Acurácia da Regressão Logística foi: {acuracia*100:.2f}%")

```

```python
output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': previsoes})
output.to_csv('submission.csv', index=False)
print("A submissão foi realizada com sucesso!")
```
