```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def criar_nuvem(texto):
    
    #Retirar palavras que não são relevantes para a análise
    stopwords = ['da','de','um', 'uma','a', 'o', 'para', 'em', 'que',"e","ou"]
    
    # Criando a nuvem de palavras
    nuvem = WordCloud(width=800, height=800, background_color='black', stopwords=stopwords).generate(texto)
    
    # Exibindo a nuvem de palavras
    plt.figure(figsize=(8,8))
    plt.imshow(nuvem)
    plt.axis('off')
    plt.show()
    
criar_nuvem('''Atualidade refere-se ao estado ou condição de estar atual, ou seja, estar em sintonia com o presente. Em outras palavras, a atualidade é o conjunto de eventos, acontecimentos e condições que estão ocorrendo agora, ou que são relevantes para o momento presente.Na sociedade atual, a tecnologia está avançando rapidamente e desempenhando um papel cada vez mais importante na vida cotidiana das pessoas. A pandemia da COVID-19 também tem sido um fator importante na atualidade, afetando a saúde, a economia e o comportamento das pessoas em todo o mundo.

Além disso, a política e a economia mundial também estão em constante mudança e evolução, e questões como mudanças climáticas, desigualdade social e direitos humanos continuam sendo desafios importantes que enfrentamos atualmente.

Em resumo, a atualidade é um termo amplo que se refere ao conjunto de eventos e condições que estão ocorrendo no momento presente, e que têm impacto na vida das pessoas em diferentes áreas, incluindo tecnologia, saúde, política e economia.''')
```
